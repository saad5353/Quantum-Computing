# Shor's Algorithm

## Goal

Factor a composite integer N.

Example:

15

Result:

3 × 5

---

## Components Used

- Superposition
- Modular Arithmetic
- Quantum Fourier Transform
- Phase Estimation
- Order Finding

---

## Why Important?

Shor's Algorithm can factor large integers exponentially faster than known classical algorithms.

This threatens RSA cryptography.

```python
import numpy as np
import matplotlib.pyplot as plt

from math import gcd

N = 15
a = 2

print("="*60)
print("SHOR'S ALGORITHM")
print("="*60)

print()

print("Target Number:",N)

print()

print("Randomly Chosen a:",a)
```

```python
# Verify Coprime

g = gcd(a,N)

print()

print("gcd(a,N) =",g)
```

```python
# Modular Exponentiation

values = []

for k in range(1,20):

    value = pow(a,k,N)

    values.append(value)

    print(
        f"{a}^{k} mod {N} = {value}"
    )

plt.figure(figsize=(8,4))

plt.plot(
    range(1,20),
    values,
    marker='o'
)

plt.title(
    f"{a}^k mod {N}"
)

plt.xlabel("k")

plt.ylabel("Value")

plt.grid()

plt.show()
```

```python
# Find Order

for r in range(1,N):

    if pow(a,r,N)==1:

        order = r
        break

print()

print("Order Found:",order)
```

```python
# Check Conditions

if order % 2 == 0:
    print("Even order ✓")
else:
    print("Odd order ✗")
```

```python
# Compute a^(r/2)

half_power = pow(
    a,
    order//2
)

print()

print("a^(r/2) =",half_power)
```

```python
# Factor Recovery

factor1 = gcd(
    half_power-1,
    N
)

factor2 = gcd(
    half_power+1,
    N
)

print()

print("Factor 1 =",factor1)

print("Factor 2 =",factor2)
```
